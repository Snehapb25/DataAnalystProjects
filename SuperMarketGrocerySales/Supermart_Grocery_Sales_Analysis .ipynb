```python
df['month']=df['Order Date'].dt.month
```

```python
category_sales = df.groupby('Category')['Sales'].sum()
category_sales
```

```python
category_sales_sorted=category_sales.sort_values(ascending=False)
category_sales_sorted.plot(kind='bar', figsize=(8,5),color='orange')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.savefig('Total_Sales_by_Category.png', dpi=300, bbox_inches='tight')
plt.show()
```

 Total Sales by Category 
This bar chart represents total sales across different product categories. Eggs, Meat & Fish emerged as the highest-selling category, followed closely by Snacks. Oil & Masala recorded comparatively lower sales. The chart highlights customer preference towards essential and high-demand food items.ries.


```python
df['Month'] = df['Order Date'].dt.month
df['Year'] = df['Order Date'].dt.year
```

```python
df[['Order Date', 'Month', 'Year']].head()
```

```python
monthly_sales = df.groupby('Month')['Sales'].sum()
monthly_sales
```

```python
import calendar

monthly_sales_named = monthly_sales.copy()
monthly_sales_named.index = [calendar.month_name[i] for i in monthly_sales_named.index]

monthly_sales_named
```

```python
monthly_sales_named.plot(kind='line', marker='o', figsize=(8,5), color='green')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.savefig('Monthly_Sles_Trend.png', dpi=300, bbox_inches='tight')
plt.show()
```

Monthly Sales Trend
This line chart shows monthly sales trends after extracting the month from the corrected Order Date. Sales are lowest during January and February, indicating lower consumer spending. A sharp increase is observed from September onwards, with peak sales in November, suggesting strong festive season demand.

```python
city_sales = df.groupby('City')['Sales'].sum()
city_sales.head()
```

```python
top_5_cities = city_sales.sort_values(ascending=False).head(5)
top_5_cities
```

```python
plt.figure(figsize=(8,5))

plt.bar(top_5_cities.index, top_5_cities.values, color='purple')

plt.title('Top 5 Cities by Sales')
plt.xlabel('City')
plt.ylabel('Sales')
plt.xticks(rotation=45)

# Add value labels on bars
for x, y in zip(top_5_cities.index, top_5_cities.values):
    plt.text(x, y, f'{y:,.0f}', ha='center', va='bottom', fontsize=9)
    plt.savefig('top_5_cities_sales.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(
    df['Discount'],
    df['Profit'],
    alpha=0.3,
    color='teal'
)
plt.title('Profit vs Discount')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.grid(True)
plt.savefig('profit_vs_discount.png', dpi=300, bbox_inches='tight')
plt.show()
```

Profit vs Discount 
This scatter plot shows the relationship between discount and profit for individual orders.
Each dot represents a single transaction with a specific discount and corresponding profit
The vertical alignment of points indicates that discounts are applied in fixed levels

Overall, the plot shows a negative relationship between discount and proit.
As discount levels increase, profit margins tend to decrease.
While discounts help attract customers, offering higher discounts reduces the profit arned
from each order, highlighting the need for an optimal discount strategy.
